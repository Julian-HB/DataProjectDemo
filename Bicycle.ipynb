```python
# Pandas-Bibliothek wird importiert und in der Variablen 'pd' abgespeichert.
import pandas as pd

# Die Funktion 'pandas.read_csv' wird verwendet, um eine CSV-Datei in ein sogenanntes DataFrame zu laden, welches eine zweidimensionale Datenstruktur mit beschrifteten Achsen darstellt.
# Daten werden aus der CSV_Datei 'nyc-east-river-bicycle-counts.csv' ausgelesen und indas Panads DataFrame 'bicycle_table' geladen.
bicycle_table = pd.read_csv('nyc-east-river-bicycle-counts.csv')

# Der Inhalt des Dataframes wird 'bicycle_table' wird ausgegeben.
bicycle_table 
```

```python
# Das Attribut 'pandas.DataFrame.loc' ermöglicht es bestimmte Zeilen und Spalten anhand deren Eigenschaften anzusprechen.
# Ueberpruefung ob eine Zeiele mit den selben Wert ('2016-04-30 00:00:00') mehrmahls vorkommt.
bicycle_table.loc[(bicycle_table['Date']=='2016-04-30 00:00:00')]
```

```python
# Die Eigenschaft bicycle_table.shape stell die Dimension des Datenrahmens(Dataframe) dar.
# Die Ausgabe er gibt 30 Zeilen und 11 Spalten: (30, 11)
bicycle_table.shape
```

```python
# Die Methode 'pandas.DataFrame.dropna' entfernt alle Zeilen 
# mit fehlenden Werten.
# Zeilen mit fehlenden Werten werden aus dem Dataframe geloescht.
bicycle_table.dropna(inplace=True)

# Die Methode 'pandas.DataFrame.drop_duplicates' entfernt doppelte Zeilen aus dem Dataframe.
# Duplikate basierend auf der Spalte 'Date' im DataFrame 
# 'bicycle_table' werden geloescht.
bicycle_table.drop_duplicates(subset='Date', inplace=True)
```

```python
# Liste von Spalten, die aus dem DataFrame 'bicycle_table' entfernt werden sollen.
columns_drop = ['Unnamed: 0', 'Day', 'Brooklyn Bridge', 'Williamsburg Bridge', 'Queensboro Bridge', 'Total']

# Die Methode 'pd.DataFrame.drop' wird verwendet, um Zeilen oder Spalten anhand von Bezeichnungen zu entfernen. 
# Liste der Spalten aus der Variable 'columns_drop' werden aus dem DataFrame 'bicycle_table' geloescht.
bicycle_table.drop(columns=columns_drop, axis=1, inplace=True)
```

```python
# Die Funktion 'pd.to_datetime' konvertiert ein Skalar, eine Array-aehnliche Struktur, 
# eine Series oder ein DataFrame/dict-aehnliches Objekt in ein Pandas-Datumszeitobjekt
# Die Datenreihe bicycle_table['Date'] wird in ein Pandas-Datetime-Objekt convertiert 
bicycle_table['Date'] = pd.to_datetime(bicycle_table['Date'])
# Pandas-Datetime-Objekt wird in ein angepasstes Stringformat(deutsche Datumsdarstellung) unkonvertiert.
bicycle_table['Date'] = pd.to_datetime(bicycle_table['Date'], format='%d/%m/%Y')
# Die Datenreihe bicycle_table['Date'] wird in ein Pandas-Datetime-Objekt convertiert.
bicycle_table['Date'] = pd.to_datetime(bicycle_table['Date'])

# Mit der Methode 'pandas.DataFrame.rename' ist es moeglich mittels einer 'Dictionary' Spalten und Indexbezeichnungen umzubennen.
# Die Spalten 'High Temp (°F)' und 'Low Temp (°F)' in 'High Temp (°C)' und 'Low Temp (°C)' werden umbennant.
bicycle_table.rename(columns={'High Temp (°F)': 'High Temp (°C)', 'Low Temp (°F)': 'Low Temp (°C)'}, inplace=True)

# Durch das ansprechen einer Spalte, ist es in Pandas moeglich jeden Werte der Spalte eines Dataframes anhand einer einzigen Operation zu bearbeiten. 
# Die Spalten, High Temp (°C) und Low Temp (°C), werden von Fahrenheit in Celsius umgerechnet.
bicycle_table['High Temp (°C)'] -= 32
bicycle_table['Low Temp (°C)'] -= 32
bicycle_table['High Temp (°C)'] *= 5/9
bicycle_table['Low Temp (°C)'] *= 5/9
```

```python
# Die Durchschnittstemperatur wird berechnet und in eine neue Spalte 'Average Temp' eingefuegt.
bicycle_table['Avarage Temp (°C)'] = bicycle_table['High Temp (°C)'] + bicycle_table['Low Temp (°C)']
bicycle_table['Avarage Temp (°C)'] /= 2
```

```python
# Die Fahrradüberquerungen der Manhattan Bridge werden kategorisiert.
# Die Kategorisierung erfolgt nach den durchschnittlichen Überquerungen(6200) der Manhattan Bridge pro Tag innerhalb eines Jahres(2018).
# Quelle: https://www.nyc.gov/html/dot/html/infrastructure/manhattan-bridge.shtml

# Die Spalte 'Bicycle Traffic Density' wird erzeugt.
bicycle_table['Bicycle Traffic Density'] = 'average'

# Das Attribut 'pandas.DataFrame.loc' ermöglicht es bestimmte Zeilen und Spalten anhand deren Eigenschaften anzusprechen.
# Die Kategorisierung erfolgt nach dem durchscnitteswert '6200'
# Unterschreitet der Wert der Spalte Manhattan Bridge diesen Wert(6200), so ist dieser Wert als unterdurchscnittlich(eng: below average) zu erachten.
bicycle_table.loc[bicycle_table['Manhattan Bridge'] <=  6200, 'Bicycle Traffic Density'] = 'below average'
# Überschreitet der Wert der Spalte Manhattan Bridge diesen Wert(6200), so ist dieser Wert als überdurchscnittlich((eng: above average)) zu erachten.
bicycle_table.loc[bicycle_table['Manhattan Bridge'] >=  6200, 'Bicycle Traffic Density'] = 'above average'
# Die ganze Tabelle wird auf zwei Nachkommmatellen gerundet um eine bessere Darstellung zu ermöglichen.
bicycle_table = bicycle_table.round(2)
```

```python
# Die Spalte 'Above Average Classification' soll die Klassiefizierungswerte als 1 oder 0 darstellen.
bicycle_table['Above Average Classification'] = 1

# Das Attribut 'pandas.DataFrame.loc' ermöglicht es bestimmte Zeilen und Spalten anhand deren Eigenschaften anzusprechen.
# Unterdurchschnittliche Werte sollen mit der Zahl 0 abgespeichert werden.
bicycle_table.loc[bicycle_table['Bicycle Traffic Density'] == 'below average', 'Above Average Classification'] = 0

# Das Attribut 'pandas.DataFrame.loc' ermoeglicht es bestimmte Zeilen und Spalten anhand deren Eigenschaften anzusprechen.
# Überdurchschnittliche Werte sollen mit der Zahl 1 abgespeichert werden.
bicycle_table.loc[bicycle_table['Bicycle Traffic Density'] == 'above average', 'Above Average Classification'] = 1
```

```python

# Die Methode 'pandas.DataFrame.drop' entfernt Zeilen oder Spalten mit einer bestimmten Bezeichnung.
# Der Niederschlag wird hearausgefiltert. Der Niederschlag soll für die betrachteten Zeilen gleich sein.
# Somit werden die Zeilen heraugeloescht welche 'Precipitation'] != '0' entsprechen.
bicycle_table.drop(bicycle_table[bicycle_table['Precipitation'] != '0'].index, inplace=True)
# Die Funktion sort_values wird verwendet, um nach den Werten entlang einer Achse(Zeilen oder Spalten) zu sortieren. 
# Der DataFrame 'bicycle_table' wird nach der Spalte 'Manhattan Bridge' in aufsteigender Reihenfolge sortiert.
bicycle_table.sort_values(by='Avarage Temp (°C)', ascending=True, inplace=True)
bicycle_table
```

```python
# Das Attribut 'pandas.DataFrame.loc' ermöglicht es bestimmte Zeilen und Spalten anhand deren Eigenschaften anzusprechen.
# Der Kategorie 'above average' aus der Zeile 'Bicycle Traffic Density' wird in der Zeile 'Above Average Classificatio' die Zahl 1 zugewiesen.
above_average = bicycle_table.loc[bicycle_table['Above Average Classification'] == 1]

# Das Attribut 'pandas.DataFrame.loc' ermöglicht es bestimmte Zeilen und Spalten anhand deren Eigenschaften anzusprechen.
# Der Kategorie 'below average' aus der Zeile 'Bicycle Traffic Density' wird in der Zeile 'Above Average Classificatio' die Zahl 0 zugewiesen.
below_average = bicycle_table.loc[bicycle_table['Above Average Classification'] == 0]
```

```python
# Die Bibliothek Numpy wird importiert.
import numpy as np


# Die Spalte 'Avarage Temp (°C)' wird als eigenes Dataframe in der variable x abgespeichert.
x = pd.DataFrame(bicycle_table['Avarage Temp (°C)'])
# Die Spalte 'Manhattan Bridge' wird als eigenes Dataframe in der variable y abgespeichert.
y = pd.DataFrame(bicycle_table['Manhattan Bridge'])

# Die Eigenschaft 'pandas.DataFrame.values' gibt einen Numpy-Array zurück.
# Die Variable x wird in einen Numpy-Array konvertiert.
x = x.values
# Die Vraribale y wird in einen Numpy-Array konvertiert.
y = y.values
```

```python
# Im Modul 'sklearn.linear_model' sind eine Reihe von linearen Modellen implementiert.

# Die Klasse 'sklearn.linear_model.LinearRegression' passt ein lineares Modell mit Koeffizienten an, 
# um die Restsumme der Quadrate zwischen den beobachteten Zielen im Datensatz 
# und den von der linearen Approximation vorhergesagten Zielen zu minimieren.

# Es wird die Klasse 'sklearn.linear_model.LinearRegression' importiert.
from sklearn.linear_model import LinearRegression

# Die Methode 'LinearRegression().fit(x, y)' passt ein lineares Modell an die Daten an. Die Koeaffizienten werden berechnet.
linear_reg = LinearRegression().fit(x, y)

# Das Attribut 'coef_' liefert ein Array von Koeffizienten, die durch die Methode 'LinearRegression().fit(x, y)' bestimmt wurden. 
# Die Anzahl der Koeffizienten (Parameter) entspricht der Anzahl der übergebenen Inputvariablen eines Modells.
# Die Steigung des Zweidimensionalen Modells wird in der Variable 'slope' abgespeicher.
slope = linear_reg.coef_
# Das Attribut 'intercept_' liefert ein Array oder einen Float von Koeffizienten, die durch die Methode 'LinearRegression().fit(x, y)' bestimmt wurden. 
# Der Y-Achsenabschnitt des Zweidimensionalen Modells wird in der Variable 'intercept' abgespeicher.
intercept = linear_reg.intercept_

# Die Steigung wird in der Variablen 'a' abgespeichert und auf zwei Nachkommastellen gerundet.
# Die Variable 'slope' ist ein 'Numpy-Array' mit einer Dimension für die Zeilen eines Datesatzes und einer Dimension für die Spalten des Datensatzes.
# Auf die Zeilen des Arrays werden mit den ersten Index und die Spalten mit dem zweiten Index zugegriffen.
a = slope[0][0].round(2)

# Der Y-Achsenabschnitt wird in der Variablen 'b' abgespeichert und auf zwei Nachkommstellen gerundet.
# Die Variable 'intercept' ist ein 'Numpy-Array' mit einer Dimension.
b = intercept[0].round(2)

# Die Methode 'LinearRegression().predict(x)' trifft vorhersagen für die uebergebenen Daten der Inputvariable.
# Die vorhergesagten Werte werden in der Variable 'y_pred' abgespeichert.
# LetztMis n Zeilen und einer Spalte.
y_pred = linear_reg.predict(x)

# Die Methode 'LinearRegression().score(X,y)' gibt das Bestimmtheitsmaß R² zurueck.
# Das Bestimmtheitsmaß wird in der Variablen 'linear_score' gespeichert und mit 'print(linaer_score)' augegeben.
linear_score = linear_reg.score(x, y).round(2)
print('R² beträgt: ',linear_score) # >>> R² beträgt:  0.78

# Die Form der linearen Funktion wird ausgegeben.
print(f"Lineare Funktion: y = {a}x + {b}")
# >>> Lineare Funktion: y = 232.41x + 2192.02

# Die Methode 'numpy.shape()' gibt einen Tupel von Integer-Werten zurück welche die Länge der Dimensionen des Arrays erfassen.
# Die Diemensionen (Zeilen und Spalten) von 'y_pred' werden ausgegeben.
print(y_pred.shape) # >>> (17, 1)
```

```python
# Die Methode 'matplotlib.pyplot.scatter' wird verwendet, um ein Streudiagramm von y gegen x mit variierender Markergroeße und/oder -farbe zu erstellen.
# Die Temperatur wird der Anzahl der Überquerungen der Manhattan Bridge in einem Streudiagramm gegenuebergestellt.
# Die Resultierenden Datenpunkte werden als hellgruene Punkte ausgegeben (color='lightgreen').
plt.scatter(bicycle_table['Avarage Temp (°C)'], bicycle_table['Manhattan Bridge'], color='lightgreen', label='Data')

# Die Funktion 'matplotlib.xlabel' in Matplotlib wird verwendet, um die Beschriftung der x-Achse in einem Diagramm festzulegen.
# Die X-Achse wird als 'Avarage Temp (°C)' beschriftet.
plt.xlabel('x-axis: Avarage Temp (°C)')

# Die Funktion 'matplotlib.xlabel' in Matplotlib wird verwendet, um die Beschriftung der Y-Achse in einem Diagramm festzulegen.
# Die Y-Achse wird als 'Bicycle Crossings per day' beschriftet.
plt.ylabel('y-axis: Bicycle Crossings per day')

# Die Funktion 'matplotlib.title' wird verwendet, um einen Titel über einem Diagramm zu platzieren.
# Der Diagramm-Titel wird gesetzt.
plt.title('Manhattan Bridge (Linear Regression)')


# Die Methode 'matplotlib.pyplot.plot' wird verwendet, um ein Linieniagramm von y zu x mit variierender Markergröße und/oder -farbe zu erstellen.
# Die Temperatur wird der Anzahl der Überquerungen der Manhattan Bridge in einem Lininien-Diagramm gegenübergestellt.
plt.plot(x, y_pred, color='black', linewidth=2, label= f"Regression Plot: y = {a} * x + {b}")

# Die Methode 'matplotlib.pyplot.legend' wird verwendet um eine Legende auf der Achse zu platzieren.
# In der Legende werden die Zusatzinformationen('labels', 'color', 'marker') zu den Diagrammen wiedergegeben.
plt.legend()

# Die Funktion 'matplotlib.show()' in Matplotlib wird verwendet, um alle offenen Diagramme anzuzeigen. 
# Der Punktdiagramm und Lininien-Diagramm  werden angezeigt.
plt.show()
```

```python
# Logistische Funktion wird definiert
def logistic_function(a, b, x, y_max):
    """
    Berechnet die logistische Funktion für gegebene Parameter.

    Parameters:
    - a: Steigung der logistischen Funktion
    - b: Verschiebung der logistischen Funktion
    - x: Eingabevariablen
    - y_max: Obere Schranke der logistischen Funktion

    Returns:
    - arr: Ausgabe der logistischen Funktion
    """
    arr = (y_max)/(1 + np.exp(a*x + b))
    return arr

# Die Methode 'numpy.max(y)' gibt das Maximum eines Arrays oder entlang einer Achse zurück.
# Die Obere Schranke für die logistische Funktion wird gesetzt
y_max = np.max(y)

# Die Methode 'numpy.log(y)' bildet für alle Elemente eines Numpy-Arrays den natürlichen logarithmischen Wert.
# Das Ergebniss dieser Logarithmierung wird in der Variablen 'y_log' gespeichert.
y_log = np.log(((y_max + 1)/y)-1)

# Die Methode 'LinearRegression().fit(x, y)' passt ein lineares Modell an die Daten an. Die Koeaffizienten werden berechnet.
# Die Variable 'log_linear_reg' representiert das Modell welches anhand von den x-Werten sowie y_log-Werten trainiert wurde.
log_linear_reg = LinearRegression().fit(x, y_log)

# Die Methode 'LinearRegression().predict(x)' trifft vorhersagen für die übergebenen Daten der Inputvariable für das trainierte Modell.
y_log_pred = log_linear_reg.predict(x)

# Das Attribut 'coef_' liefert ein Array von Koeffizienten, die durch die Methode 'LinearRegression().fit(x, y)' bestimmt wurden.
# Die Steigung von dem Modell 'log_linear_reg' wird  in der Variablen 'a_slope' abgespeichert.
a_slope = log_linear_reg.coef_[0][0].round(2)

# Das Attribut 'intercept_' liefert ein Array oder einen Float von Koeffizienten, die durch die Methode 'LinearRegression().fit(x, y)' bestimmt wurden.
# Der y-Achsenabschnitt von dem Modell 'log_linear_reg' wird  in der Variablen 'b_intercept' abgespeichert.
b_intercept = log_linear_reg.intercept_[0].round(2)

# Die Methode 'numpy.linspace' gibt eine Reihe von gleichmäßig verteilten Zahlen über ein angegebenes Intervall zurück.
# Die Methode 'numpy.linspace' wird angewendet um einen gleichmäßigen Funktionsverlauf innerhalb des Graphen von der Bibliothek 'matplotlib.pyplot' darzustellen.
# Die Varibale 'x_values' wird verwendet um den Werteverlauf der logistischen Funktion mögliochst genau abzubilden.
x_values = np.linspace(-20, 40)

# Die logistischen Funktionswerte für die gegebenen x-Werte werden mittels der zuvor definierten Funktion 'logistic_function' vorhergesagt.
y_log_pred_end = logistic_function(a_slope, b_intercept, x_values, y_max)
```

```python
# Die Methode 'matplotlib.pyplot.scatter' wird verwendet, um ein Streudiagramm von y gegen x mit variierender Markergröße und/oder -farbe zu erstellen.
# Die Temperatur wird der Anzahl der Überquerungen der Manhattan Bridge in einem Streudiagramm gegenübergestellt.
# Die Resultierenden Datenpunkte werden als hellgrüne Punkte ausgegeben (color='lightgreen').
plt.scatter(bicycle_table['Avarage Temp (°C)'], bicycle_table['Manhattan Bridge'], color='lightgreen', label='Data')

# Die Funktion 'matplotlib.xlabel' in Matplotlib wird verwendet, um die Beschriftung der x-Achse in einem Diagramm festzulegen.
# Die X-Achse wird als 'Avarage Temp (°C)' beschriftet.
plt.xlabel('x-axis: Avarage Temp (°C)')

# Die Funktion 'matplotlib.xlabel' in Matplotlib wird verwendet, um die Beschriftung der Y-Achse in einem Diagramm festzulegen.
# Die Y-Achse wird als 'Bicycle Crossings per day' beschriftet.
plt.ylabel('y-axis: Bicycle Crossings per day')

# Die Funktion 'matplotlib.title' wird verwendet, um einen Titel über einem Diagramm zu platzieren.
# Der Diagramm-Titel wird gesetzt.
plt.title('Manhattan Bridge (Logistic Regression)')

# Die Methode 'matplotlib.pyplot.plot' wird verwendet, um ein Linieniagramm von y zu x mit variierender Markergröße und/oder -farbe zu erstellen.
# Die Temperatur wird der Anzahl der Überquerungen der Manhattan Bridge in einem Lininien-Diagramm gegenübergestellt.
plt.plot(x_values, y_log_pred_end, color='black', label=fr' Regression Plot: $y = \frac{{{y_max}}}{{1 + e^{{{a_slope}x + {b_intercept}}}}}$')

# Die Methode 'matplotlib.pyplot.legend' wird verwendet um eine Legende auf der Achse zu platzieren.
# In der Legende werden die Zusatzinformationen('labels', 'color', 'marker') zu den Diagrammen wiedergegeben.
plt.legend(loc='upper left', fontsize='large')

# Die Funktion 'matplotlib.show()' in Matplotlib wird verwendet, um alle offenen Diagramme anzuzeigen. 
# Der Punktdiagramm und Lininien-Diagramm  werden angezeigt..
plt.show()
```

```python
# Exponentielle Funktion wird definiert
def exponential_function(a, b, x):
    """
    Berechnet die exponentielle Funktion für gegebene Parameter.

    Parameters:
    - a: Basis der exponentiellen Funktion
    - b: Skalierungsfaktor der exponentiellen Funktion
    - x: Eingabevariablen

    Returns:
    - exp_arr: Ausgabe der exponentiellen Funktion
    """
    exp_arr = np.power(a, x)
    exp_arr *= b
    return exp_arr

# Die Methode 'numpy.log10(y)' berechnet den dekadischen Logarithmus für alle Elemente eines Numpy-Arrays.
# Das Ergebniss dieser Logarithmierung wird in der Variablen 'y_log_exp' gespeichert.
y_log_exp = np.log10(y)

# Die Methode 'LinearRegression().fit(x, y)' passt ein lineares Modell an die Daten an. Die Koeaffizienten werden berechnet.
# Die Variable 'exp_linear_reg' representiert das Modell welches anhand von den Werten der Variablen 'x' sowie 'y_log_exp' trainiert wurden.
exp_linear_reg = LinearRegression().fit(x, y_log_exp)

# Die Methode 'LinearRegression().predict(x)' trifft vorhersagen für die übergebenen Daten der Inputvariable für das trainierte Modell
#y_exp_pred = exp_linear_reg.predict(x)

# Das Attribut 'coef_' liefert ein Array von Koeffizienten, die durch die Methode 'LinearRegression().fit(x, y)' bestimmt wurden.
# Die Steigung von dem Modell 'exp_linear_reg' wird  in der Variablen 'exp_log_slope' abgespeichert.
exp_log_slope = exp_linear_reg.coef_

# Das Attribut 'intercept_' liefert ein Array oder einen Float von Koeffizienten, die durch die Methode 'LinearRegression().fit(x, y)' bestimmt wurden.
# Der y-Achsenabschnitt von dem Modell 'exp_linear_reg' wird  in der Variablen 'exp_log_intercept' abgespeichert.
exp_log_intercept = exp_linear_reg.intercept_

# Die Funktion numpy.power gibt die Zahl des ersten Elements, potenziert mit der Zahl des zweiten Elements, zurück. 
# Umrechnung der logarithmischen Werte in exponentielle Werte
a_exp_slope = np.power(10, exp_log_slope[0][0]).round(2)
b_exp_intercept = np.power(10, exp_log_intercept[0]).round(2)

# Die Methode 'numpy.linspace' gibt eine Reihe von gleichmäßig verteilten Zahlen über ein angegebenes Intervall zurück.
# Die Methode 'numpy.linspace' wird angewendet um einen gleichmäßigen Funktionsverlauf innerhalb des Graphen von der Bibliothek 'matplotlib.pyplot' darzustellen.
# Die Varibale 'x_values' wird verwendet um den Werteverlauf der logistischen Funktion mögliochst genau abzubilden.
x_values = np.linspace(0, 20)

# Die Funktionswerte für die gegebenen x-Werte der exponentiellen Funktion werden mittels der zuvor definierten Funktion 'exponential_function(a, b, x):' vorhergesagt.
y_exp_pred_end = exponential_function(a_exp_slope, b_exp_intercept, x_values)
```

```python
# Potenzfunktion wird definiert
def power_function(a, b, x):
    """
    Berechnet die Potenzfunktion für gegebene Parameter.

    Parameters:
    - a: Exponent der Potenzfunktion
    - b: Skalierungsfaktor der Potenzfunktion
    - x: Eingabevariablen

    Returns:
    - exp_arr: Ausgabe der Potenzfunktion
    """
    exp_arr = np.power(x, a) * b
    return exp_arr

# Die Methode 'numpy.log10(y)' berechnet den dekadischen Logarithmus für alle Elemente eines Numpy-Arrays.
# Das Ergebniss dieser Logarithmierung wird in der Variablen 'y_log_pow' gespeichert.
y_log_pow = np.log10(y)

# Das Ergebniss dieser Logarithmierung wird in der Variablen 'x_log_pow' gespeichert.
x_log_pow = np.log10(x)

# Die Methode 'LinearRegression().fit(x, y)' passt ein lineares Modell an die Daten an. Die Koeaffizienten werden berechnet.
# Die Variable 'power_linear_reg' representiert das Modell welches anhand von den Werten der Variablen 'x_log_pow' sowie 'y_log_pow' trainiert wurde.
power_linear_reg = LinearRegression().fit(x_log_pow, y_log_pow)

# Das Attribut 'coef_' liefert ein Array von Koeffizienten, die durch die Methode 'LinearRegression().fit(x, y)' bestimmt wurden.
# Die Steigung von dem Modell 'power_linear_reg' wird  in der Variablen 'a_power_slope' abgespeichert.
a_power_slope = power_linear_reg.coef_[0][0].round(2)

# Das Attribut 'intercept_' liefert ein Array oder einen Float von Koeffizienten, die durch die Methode 'LinearRegression().fit(x, y)' bestimmt wurden.
# Der y-Achsenabschnitt von dem Modell 'power_linear_reg' wird  in der Variablen 'b_power_intercept' abgespeichert.
b_power_intercept = np.power(10, power_linear_reg.intercept_[0]).round(2)

# Die Methode 'numpy.linspace' gibt eine Reihe von gleichmäßig verteilten Zahlen über ein angegebenes Intervall zurück.
# Die Methode 'numpy.linspace' wird angewendet um einen gleichmäßigen Funktionsverlauf innerhalb des Graphen von der Bibliothek 'matplotlib.pyplot' darzustellen.
# Die Varibale 'x_values' wird verwendet um den Werteverlauf der logistischen Funktion mögliochst genau abzubilden.
x_values = np.linspace(0, 30)

# Die Funktionswerte für die gegebenen x-Werte der Potenzfunktion werden mittels der zuvor definierten Funktion 'power_function(a, b, x):' vorhergesagt.
y_power_pred_end = power_function(a_power_slope, b_power_intercept, x_values)
```

```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# Annahme: x und y sind bereits definierte Daten.
X = np.array(x)
Y = np.array(y)

# Der 'K-Folds cross-validator' wird benutzt, um den Datensatz in k (Anzahl) Teil-Datensätze zu unterteilen.
# Das Argument 'n_splits=4' weist auf die Anzahl der Teildatensätze hin.
kf = KFold(n_splits=4)

# Die Methode 'KFold.get_n_splits(X)' gibt die Anzahl der Aufteilungsiterationen im Kreuzvalidierer zurück.
splits = kf.get_n_splits(X)
print('Die Anzahl der Teildatensätze: ', splits)

# Der Array 'model_container' wird zum Speichern der berechneten Modelle verwendet.
model_container = []

# for-Schleife zum Iterieren der Teildatensätze.
# kf.split(X) generiert Indizes, um die Teildatensätze in Test- und Trainingsdatensätze zu unterteilen.
for i, (train_index, test_index) in enumerate(kf.split(X)):
    print('Das ist das Trainingsset: ',train_index)
    print('Das ist das Testset: ',test_index)
    
    # Die Trainingsdaten werden mittels der Indizes bestimmt.
    x_train, y_train = X[train_index], Y[train_index]
    
    # Die Testdaten werden mittels der Indizes bestimmt.
    x_test, y_test = X[test_index], Y[test_index]
    
    # Das lineare Modell wird erstellt.
    linear_model = LinearRegression().fit(x_train.reshape(-1, 1), y_train)
    
    # Steigung (slope) und y-Achsenabschnitt (intercept) werden aus der linearen Regression extrahiert.
    slope = linear_model.coef_[0][0].round(2)
    intercept = linear_model.intercept_[0].round(2)
    
    # Vorhersage für die Testdaten wird berechnet.
    # Die Methode 'numpy.reshape' gibt einem Array eine neue Form, ohne die Daten zu ändern.
    # Der erste Parameter innerhalb von 'numpy.reshape(-1,1)' gibt die Zeilenanzahl an und das zweite Argument gibt die Spaltenanzahl an.
    # Die Zahl '-1' gibt innerhalb von 'numpy.reshape(-1,1)' eine unbestimmte Dimension an.
    y_pred = linear_model.predict(x_test.reshape(-1, 1))
    
    # Das Bestimmtheitsmaß (r2_val) wird berechnet.
    #r2_val = r2_score(y_train, y_pred)
    r2_val = 1 - (((y_test - y_pred)** 2).sum() / ((y_test - y_train.mean()) ** 2).sum())
    print('Das ist R²: ', r2_val)
    
    # Das Dictionary 'num_dict' speichert die wichtigsten Elemente des trainierten Modells.
    num_dict = {'r2': r2_val, 'slope': slope, 'intercept': intercept, 'x_test': x_test, 'y_test': y_test, 'x_train': x_train, 'y_train': y_train}
    
    # Der Array 'model_container' speichert 'num_dict', um die berechneten Modelle miteinander zu vergleichen.
    model_container.append(num_dict)
    

    # Der Numpy-Array hat eine Reihe von generierten x-Werten (Temperaturwerten).
    x_values = np.linspace(min(X), max(X), 100).reshape(-1, 1)
    # Der Numpy-Array 'y_pred_end' speichert die Funktionswerte aus dem linearen Modell.
    y_pred_end = num_dict['slope'] * x_values + num_dict['intercept']
    
    # Streudiagramme werden erstellt.
    plt.scatter(x_train, y_train, color='lightgreen', label='Training Data')
    plt.scatter(x_test, y_test, color='#ff6b28', label='Test Data')

    # Labels und Titel werden gesetzt.
    plt.xlabel('Average Temp (°C)')
    plt.ylabel('Bicycle Crossings per day')
    plt.title('Manhattan Bridge (Linear Regression)')

    # Liniendiagramm wird hinzugefügt.
    plt.plot(x_values, y_pred_end, color='black', label= f"Regression Plot: y = {slope} * x + {intercept}")
    
    plt.legend()            
                
    # Diagramme werden angezeigt.
    plt.show()

# Der Array 'model_container' wird mit der Funktion sorted() nach dem Bestimmtheitsmaß aufsteigend sortiert. 
model_container = sorted(model_container, key=lambda x: x['r2'], reverse=True)

# Das beste Modell befindet sich an der ersten Stelle.
model_one = model_container[0]['r2']

# Die Funktionswerte werden ein weiteres Mal berechnet, um eine Darstellung zu ermöglichen.
y_pred_end = model_container[0]['slope'] * x_values + model_container[0]['intercept']
```
